```python
import matplotlib.pyplot as plt


BSWN = [3271150, 3342000, 7246500]
BSUNC = [3360150, 3160250, 7072750]
BSPPI = [3180600, 3278700, 7162750]

# Positions for the groups
positions = range(3)

# Width of the bars
width = 0.2

plt.bar(positions, BSWN, width=width, label='BubbleSortWhileNeeded')
plt.bar([pos + width for pos in positions], BSUNC, width=width, label='BubbleSortUntilNoChange')
plt.bar([pos + 2 * width for pos in positions], BSPPI, width=width, label='BubbleSortPassPerItem')

plt.xticks([pos + width for pos in positions], ['Integer', 'Double', 'String'])

plt.xlabel('Case')
plt.ylabel('Execution Time')
plt.title('Data type')
plt.legend()

plt.show()
```
